### Which one should you use ? (pyplot vs matplotlib OO method?)   
* When plotting something quickly, okey to use the pyplot method
* When plotting something more advance ,use OO method 

```python
over_50 = heart_disease[heart_disease['age']>50]
```

```python
over_50
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target')
```

```python

# OO method mixed with pyplot
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim([45,100])
```

```python
# OO method from scrach
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

#Customize the plot
ax.set(title='Heart Disease and Cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol');

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line 
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
#Subplot of chol age thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)
#Add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])

#customize ax0
ax0.set(title='Heart Disease and cholesterol Levels',
       
       ylabel='Cholesterol');
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle="--")



# Add data to ax1

scatter=ax1.scatter(x=over_50['age'],
                   y=over_50['thalach'],
                   c=over_50['target'])

#customize ax1
ax1.set(title='Heart Disease and max heart rate',
       xlabel="Age",
       ylabel='Thalach');
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle="--");

#Add a title for figure
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')
# OO method from scrach
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap = 'winter')# this changes colour scheme

#Customize the plot
ax.set(title='Heart Disease and Cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol');

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line 
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
# customizing x and y limitations

#Subplot of chol age thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)
#Add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'],
                   cmap='winter')

#customize ax0
ax0.set(title='Heart Disease and cholesterol Levels',
       ylabel='Cholesterol');

#Change the x axis limit
ax0.set_xlim([50,80])

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle="--")



# Add data to ax1

scatter=ax1.scatter(x=over_50['age'],
                   y=over_50['thalach'],
                   c=over_50['target'],
                   cmap='winter')

#customize ax1
ax1.set(title='Heart Disease and max heart rate',
       xlabel="Age",
       ylabel='Thalach');

#Change the x axis limit
ax1.set_xlim([50,80])

#Change the y axis limit
ax1.set_ylim([60,200])

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle="--");

#Add a title for figure
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');
```
