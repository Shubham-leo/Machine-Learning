```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def prediction_function(age):
    z=0.12963*age -5.13774
    y=sigmoid(z)
    return y
```

```python
age = 35
prediction_function(age)
```

* 0.3541 is less than 0.5 which means person with 35 age will not buy insurance

```python
age = 43
prediction_function(age)
```

* 0.60738 is more than 0.5 which means person with 43 will buy the insurance
